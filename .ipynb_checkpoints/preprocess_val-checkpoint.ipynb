```python
# let's read & visualize a single nii image and see how it looks;
# here we'll just examine the flair image...
single_img_flair = VAL_PATH+"\BraTS20_Validation_004\BraTS20_Validation_004_flair.nii" 
single_img_tice = VAL_PATH+"\BraTS20_Validation_004\BraTS20_Validation_004_t1ce.nii"
single_img_t1 = VAL_PATH+"\BraTS20_Validation_004\BraTS20_Validation_004_t1.nii"
single_img_t2 = VAL_PATH+"\BraTS20_Validation_004\BraTS20_Validation_004_t2.nii"
# single_img_mask = VAL_PATH+"\BraTS20_Training_355\BraTS20_Training_004_seg.nii"
```

```python
imgData_flair = ni.load(single_img_flair).get_fdata()
imgData_t1 = ni.load(single_img_t1).get_fdata()
imgData_tice = ni.load(single_img_tice).get_fdata()
imgData_t2 = ni.load(single_img_t2).get_fdata()
# imgData_mask = ni.load(single_img_mask).get_fdata()
# save some memeory by converting floating point numbers to int
# in the masks datasets..

# imgData_mask = imgData_mask.astype(np.uint8)
```

```python
# scale the image pixels to between 0 and 1 
# to ensure that all images are of the same pixel length or format;

scaler = MinMaxScaler()
```

```python
# since scalers are applied to 2D arrays,
# let's convert this into 2d and then back to original image;

scale_imgFlair  = scaler.fit_transform(imgData_flair.reshape(-1, imgData_flair.shape[-1])).reshape(imgData_flair.shape)
scale_imgTice =  scaler.fit_transform(imgData_tice.reshape(-1, imgData_tice.shape[-1])).reshape(imgData_tice.shape)
scale_imgT2 = scaler.fit_transform(imgData_t2.reshape(-1, imgData_t2.shape[-1])).reshape(imgData_t2.shape)
scale_imgT1 = scaler.fit_transform(imgData_t1.reshape(-1, imgData_t1.shape[-1])).reshape(imgData_t1.shape)
```

```python
# visualize img
n_slice = random.randint(0, imgData_flair.shape[2])
plt.subplot(231)
plt.imshow(scale_imgFlair[:,:, n_slice], cmap='grey')
plt.subplot(232)
plt.imshow(scale_imgT1[:,:, n_slice], cmap='grey')
plt.subplot(233)
plt.imshow(scale_imgTice[:,:, n_slice], cmap='grey')
plt.subplot(234)
plt.imshow(scale_imgT2[:,:, n_slice], cmap='grey')
# plt.subplot(235)
# plt.imshow(imgData_mask[:,:, n_slice])
```

```python
# Now after some preprocessing, combine the training data into one volume
# Since there are 3-4 volumes/planes, we combine them into one for easy model training
# other than training all planes separately..

combine_img = np.stack([scale_imgFlair, scale_imgTice, scale_imgT2], axis=3)
```

```python
# Also, crop the none relevant portion of the images to enhance the image
# This is done because there're portions that are not relevant to be trained.

combine_img = combine_img[56:184, 56:184, 13:141]
# imgData_mask = imgData_mask[56:184, 56:184, 13:141]
```

```python
# plot the cropped images to see how it looks now..
n_slice = random.randint(0, imgData_flair.shape[2])
plt.figure(figsize=(10,6))
plt.subplot(221)
plt.imshow(combine_img[:,:, n_slice, 0], cmap='grey')
plt.subplot(222)
plt.imshow(combine_img[:,:, n_slice, 1], cmap='grey')
plt.subplot(223)
plt.imshow(combine_img[:,:, n_slice, 2], cmap='grey')
# plt.subplot(224)
# plt.imshow(imgData_mask[:,:, n_slice])
```

### Do same for validation set..

```python
# put all preprocessing into a loop...
# Read nii files for each participant using glob since there different many folders to extract from
# using glob to get the filenames and use that to iterate through the file list and read the files into memory
imgflair_list_val = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_ValidationData\MICCAI_BraTS2020_ValidationData\*\*flair.nii"))
imgt1ce_list_val = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_ValidationData\MICCAI_BraTS2020_ValidationData\*\*t1ce.nii"))
imgT2_list_val = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_ValidationData\MICCAI_BraTS2020_ValidationData\*\*t2.nii"))
# imgMask_list_val = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_ValidationData\MICCAI_BraTS2020_ValidationData\*\*seg.nii"))
```

```python
# for each img folder, read its contents, scale

for img in range(len(imgflair_list_val)):
    
    # read the img and scale..
    
    # read for flair images..
    imgflair_temp = ni.load(imgflair_list_val[img]).get_fdata()
    imgflair_temp = scaler.fit_transform(imgflair_temp.reshape(-1, imgflair_temp.shape[-1])).reshape(imgflair_temp.shape)
    
    
    # read for t1ce images..
    imgT1ce_temp = ni.load(imgt1ce_list_val[img]).get_fdata()
    imgT1ce_temp = scaler.fit_transform(imgT1ce_temp.reshape(-1, imgT1ce_temp.shape[-1])).reshape(imgT1ce_temp.shape)
    
    # read for t2 images
    imgT2_temp = ni.load(imgT2_list_val[img]).get_fdata()
    imgT2_temp = scaler.fit_transform(imgT2_temp.reshape(-1, imgT2_temp.shape[-1])).reshape(imgT2_temp.shape)
    
    
    # # read for masks..
    # imgMask_temp = ni.load(imgflair_list_val[img]).get_fdata()
    # imgMask_temp = imgMask_temp.astype(np.uint8)
    # imgMask_temp[imgMask_temp==4] = 3
    
    
    # stack img channels/bands..
    
    combine_img_temp = np.stack([imgflair_temp,imgT1ce_temp,imgT2_temp], axis=3)
    
    # crop images to remove less important feature planes..
    
    combine_img_temp = combine_img_temp[56:184, 56:184, 13:141]
    # imgMask_temp = imgMask_temp[56:184, 56:184, 13:141]
    
    # get images which has segmented volumns of at least 1%,
    # i.e., images with some form of annotation information.
    
    # vals, counts = np.unique(imgMask_temp, return_counts=True)
    
    # if found to have some segmented annotation, save the image to the new folder..
    
    # if (1- (counts[0]/counts.sum())) > 0.01:
    # imgMask_temp = to_categorical(imgMask_temp, num_classes=4)
    np.save("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_ValidationData\preprocess_valid\images_val"+str(img) + '.npy', combine_img_temp)
    # np.save("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_ValidationData\preprocess_valid\masks_val"+str(img) + '.npy', imgMask_temp)
    print('save image..')
        
    # else:
    # print('Do not save image')
```
