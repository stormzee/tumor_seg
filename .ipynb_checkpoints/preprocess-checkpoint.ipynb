```python
@tf.keras.utils.register_keras_serializable()
class CustomIOUScore(tf.keras.metrics.Metric):
    def __init__(self, threshold=0.5, name='iou_score', dtype=None):
        super().__init__(name=name, dtype=dtype)
        self.threshold = threshold
        self.iou_score = sm.metrics.IOUScore(threshold=threshold)
        self.iou = self.add_weight(name='iou', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        iou = self.iou_score(y_true, y_pred)
        self.iou.assign(iou)

    def result(self):
        return self.iou

    def reset_states(self):
        self.iou.assign(0.0)

    def get_config(self):
        config = super().get_config()
        config.update({
            "threshold": self.threshold
        })
        return config

    @classmethod
    def from_config(cls, config):
        return cls(**config)

    
    
custom_iou = CustomIOUScore(threshold=0.5)





# diceLoss = sm.losses.DiceLoss(class_weights = np.array([w0, w1, w2, w3]))
# # focaLoss = sm.losses.CategoricalFocalLoss()
# # Define a custom loss function combining Dice Loss and Focal Loss

# # dice_loss = sm.losses.DiceLoss()

# # Combine Dice Loss and Focal Loss
# def combined_loss(y_true, y_pred):
#     w0, w1, w2, w3 = 0.25, 0.25, 0.25, 0.25
#     dice = sm.losses.DiceLoss(class_weights = np.array([w0, w1, w2, w3]))
#     focal = focal_loss()(y_true, y_pred)
#     return dice + focal

# total_loss = diceLoss + (1 * focaLoss)


@tf.keras.utils.register_keras_serializable()
class CombinedLoss(tf.keras.losses.Loss):
    def __init__(self, dice_weight=1.0, focal_weight=1.0, reduction=tf.keras.losses.Reduction.SUM_OVER_BATCH_SIZE, name="combined_loss"):
        super().__init__(reduction=reduction, name=name)
        self.dice_weight = dice_weight
        self.focal_weight = focal_weight
        self.dice_loss = sm.losses.DiceLoss()
        self.focal_loss = self._focal_loss()

    def _focal_loss(self, gamma=2.0, alpha=0.25):
        def focal_loss_fixed(y_true, y_pred):
            y_true = tf.cast(y_true, tf.float32)
            y_pred = tf.cast(y_pred, tf.float32)
            
            epsilon = tf.keras.backend.epsilon()
            y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
            
            y_pred_log = tf.math.log(y_pred)
            y_true_log = tf.math.log(1. - y_pred)
            
            loss = -y_true * (alpha * tf.pow((1. - y_pred), gamma) * y_pred_log) - \
                   (1. - y_true) * ((1 - alpha) * tf.pow(y_pred, gamma) * y_true_log)
            
            return tf.reduce_mean(loss)
        return focal_loss_fixed

    def call(self, y_true, y_pred):
        dice = self.dice_loss(y_true, y_pred)
        focal = self.focal_loss(y_true, y_pred)
        return self.dice_weight * dice + self.focal_weight * focal



combined_loss = CombinedLoss(dice_weight=1.0, focal_weight=1.0)
```

```python
# put all preprocessing into a fxn...
# Read nii files for each participant using glob since there different many folders to extract from
# using glob to get the filenames and use that to iterate through the file list and read the files into memory
imgflair_list = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_TrainingData\MICCAI_BraTS2020_TrainingData\*\*flair.nii"))
imgt1ce_list = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_TrainingData\MICCAI_BraTS2020_TrainingData\*\*t1ce.nii"))
imgT2_list = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_TrainingData\MICCAI_BraTS2020_TrainingData\*\*t2.nii"))
imgMask_list = sorted(glob.glob("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_TrainingData\MICCAI_BraTS2020_TrainingData\*\*seg.nii"))
```

```python
# for each img folder, read its contents, scale

for img in range(len(imgflair_list)):
    
    # read the img and scale..
    
    # read for flair images..
    imgflair_temp = ni.load(imgflair_list[img]).get_fdata()
    imgflair_temp = scaler.fit_transform(imgflair_temp.reshape(-1, imgflair_temp.shape[-1])).reshape(imgflair_temp.shape)
    
    
    # read for t1ce images..
    imgT1ce_temp = ni.load(imgt1ce_list[img]).get_fdata()
    imgT1ce_temp = scaler.fit_transform(imgT1ce_temp.reshape(-1, imgT1ce_temp.shape[-1])).reshape(imgT1ce_temp.shape)
    
    # read for t2 images
    imgT2_temp = ni.load(imgT2_list[img]).get_fdata()
    imgT2_temp = scaler.fit_transform(imgT2_temp.reshape(-1, imgT2_temp.shape[-1])).reshape(imgT2_temp.shape)
    
    
    # read for masks..
    imgMask_temp = ni.load(imgMask_list[img]).get_fdata()
    imgMask_temp = imgMask_temp.astype(np.uint8)
    imgMask_temp[imgMask_temp==4] = 3
    
    
    # stack img channels/bands..
    
    combine_img_temp = np.stack([imgflair_temp,imgT1ce_temp,imgT2_temp], axis=3)
    
    # crop images to remove less important feature planes..
    
    combine_img_temp = combine_img_temp[56:184, 56:184, 13:141]
    imgMask_temp = imgMask_temp[56:184, 56:184, 13:141]
    
    # get images which has segmented volumns of at least 1%,
    # i.e., images with some form of annotation information.
    
    vals, counts = np.unique(imgMask_temp, return_counts=True)
    
    # if found to have some segmented annotation, save the image to the new folder..
    
    if (1- (counts[0]/counts.sum())) > 0.01:
        imgMask_temp = to_categorical(imgMask_temp, num_classes=4)
        np.save("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_TrainingData\preprocess_train\images\image_"+str(img) + '.npy', combine_img_temp)
        np.save("D:\deeplearning\deploy_brain_tumor_detector\BraTS_2020\BraTS2020_TrainingData\preprocess_train\masks\mask_"+str(img) + '.npy', imgMask_temp)
        print('save image..')
        
    else:
        print('Do not save image')
```
