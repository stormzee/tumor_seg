```python


# diceLoss = sm.losses.DiceLoss(class_weights = np.array([w0, w1, w2, w3]))
# # focaLoss = sm.losses.CategoricalFocalLoss()
# # Define a custom loss function combining Dice Loss and Focal Loss

# # dice_loss = sm.losses.DiceLoss()

# # Combine Dice Loss and Focal Loss
# def combined_loss(y_true, y_pred):
#     w0, w1, w2, w3 = 0.25, 0.25, 0.25, 0.25
#     dice = sm.losses.DiceLoss(class_weights = np.array([w0, w1, w2, w3]))
#     focal = focal_loss()(y_true, y_pred)
#     return dice + focal

# total_loss = diceLoss + (1 * focaLoss)


@tf.keras.utils.register_keras_serializable()
class CombinedLoss(tf.keras.losses.Loss):
    def __init__(self, dice_weight=1.0, focal_weight=1.0, reduction=tf.keras.losses.Reduction.SUM_OVER_BATCH_SIZE, name="combined_loss"):
        super().__init__(reduction=reduction, name=name)
        self.dice_weight = dice_weight
        self.focal_weight = focal_weight
        self.dice_loss = sm.losses.DiceLoss()
        self.focal_loss = self._focal_loss()

    def _focal_loss(self, gamma=2.0, alpha=0.25):
        def focal_loss_fixed(y_true, y_pred):
            y_true = tf.cast(y_true, tf.float32)
            y_pred = tf.cast(y_pred, tf.float32)
            
            epsilon = tf.keras.backend.epsilon()
            y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
            
            y_pred_log = tf.math.log(y_pred)
            y_true_log = tf.math.log(1. - y_pred)
            
            loss = -y_true * (alpha * tf.pow((1. - y_pred), gamma) * y_pred_log) - \
                   (1. - y_true) * ((1 - alpha) * tf.pow(y_pred, gamma) * y_true_log)
            
            return tf.reduce_mean(loss)
        return focal_loss_fixed

    def call(self, y_true, y_pred):
        dice = self.dice_loss(y_true, y_pred)
        focal = self.focal_loss(y_true, y_pred)
        return self.dice_weight * dice + self.focal_weight * focal



combined_loss = CombinedLoss(dice_weight=1.0, focal_weight=1.0)
```

```python
@tf.keras.utils.register_keras_serializable()
class CustomIOUScore(tf.keras.metrics.Metric):
    def __init__(self, threshold=0.5, name='iou_score', dtype=None):
        super().__init__(name=name, dtype=dtype)
        self.threshold = threshold
        self.iou_score = sm.metrics.IOUScore(threshold=threshold)
        self.iou = self.add_weight(name='iou', initializer='zeros')

    def update_state(self, y_true, y_pred, sample_weight=None):
        iou = self.iou_score(y_true, y_pred)
        self.iou.assign(iou)

    def result(self):
        return self.iou

    def reset_states(self):
        self.iou.assign(0.0)

    def get_config(self):
        config = super().get_config()
        config.update({
            "threshold": self.threshold
        })
        return config

    @classmethod
    def from_config(cls, config):
        return cls(**config)

    
    
custom_iou = CustomIOUScore(threshold=0.5)
```

```python
metrics_ = ['accuracy', custom_iou]
optimizer_ = tf.keras.optimizers.Adam(learning_rate=0.001)
```
